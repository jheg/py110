```python
%%timeit -n 10 -r 3
def average(list_of_integers):
    total = 0
    entries = len(list_of_integers)
    result = 0
    for number in list_of_integers:
        total += number
    
    result = total / entries

    return int(result)

print(average([1, 5, 87, 45, 8, 8]) == 25)        # True
print(average([9, 47, 23, 95, 16, 52]) == 40)     # True
print(average([7]) == 7)                          # True
print(average(list(range(1, 100000))) == 50000)
```

```python
%%timeit -n 10 -r 3
def average(numbers):
    return sum(numbers) // len(numbers)

print(average([1, 5, 87, 45, 8, 8]) == 25)        # True
print(average([9, 47, 23, 95, 16, 52]) == 40)     # True
print(average([7]) == 7)                          # True
print(average(list(range(1, 100000))) == 50000)
```
